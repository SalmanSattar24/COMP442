```python
import os
import pandas as pd

# read in the data
IRIS_PATH = "iris.csv"
def load_iris_data(iris_path=IRIS_PATH):
    csv_path = os.path.join(iris_path)
    return pd.read_csv(csv_path)

```

```python
# a) Print how many tuples the dataset has
iris = load_iris_data()
print("Number of tuples: ", len(iris))
```

```python
# g) Create four individual histogram plots in a single figure, where each plot contains one 
#   feature  (petal_width, petal_length, sepal_width, sepal_length). Hint: use Pandas library

%matplotlib inline
import matplotlib.pyplot as plt
# iris.hist(bins=50, figsize=(20,15))
# plt.show()

# Use seaborn library do the same thing
import seaborn as sns
sns.set_theme(style="darkgrid", color_codes=True)

# Plot a histogram for each feature

df = sns.load_dataset('iris')
sns.set_theme(style="white", color_codes=True)

fig, ax = plt.subplots(2, 2, figsize=(10,10))
sns.histplot(data=df, x="sepal_length", kde=True, color="red", ax=ax[0,0])
sns.histplot(data=df, x="sepal_width", kde=True, color="gold", ax=ax[0,1])
sns.histplot(data=df, x="petal_length", kde=True, color="green", ax=ax[1,0])
sns.histplot(data=df, x="petal_width", kde=True, color="blue", ax=ax[1,1])

plt.show()

# Now put all the histograms in one figure







```

```python
# h) Create four individual boxplots in a single figure, where each plot contains one 
#    feature  (petal_width, petal_length, sepal_width, sepal_length).

# iris.boxplot(by="class", figsize=(20,15))
# plt.show()

# Use seaborn library do the same thing
import seaborn as sns
sns.set(style="white", color_codes=True)
fig, ax = plt.subplots(2, 2, figsize=(10,10))
sns.boxplot(x="class", y="sepal_length", data=iris, ax=ax[0,0])
sns.boxplot(x="class", y="sepal_width", data=iris, ax=ax[0,1])
sns.boxplot(x="class", y="petal_length", data=iris, ax=ax[1,0])
sns.boxplot(x="class", y="petal_width", data=iris, ax=ax[1,1])
plt.show()

```
